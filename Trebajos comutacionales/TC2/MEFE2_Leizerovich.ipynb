```python
import numpy as np
from matplotlib import pyplot as plt
import scipy
from scipy.optimize import curve_fit
from scipy.stats import cauchy
import random
```

```python
def plot_histogram(data, bins, ticks=5, xlabel='', ylabel='', density=True, ax = None,
                   color='', ecolor='red' , alpha = 1, label = 'Histograma'):      

    N = len(data)
    num_bins = len(bins)
    xmin = min(bins)
    xmax = max(bins)
    
    
    if ax is None:
        ax = plt.gca()
    
    hist, _ = np.histogram(data, bins=bins, density=density)
    
    normed = N if density else 1
    error =  np.sqrt(hist / normed)   
    
    
    # Se le puede pasar un yerr y xerr, como a errorbar
    # ecolor es el color de los errores y capsize es el tamaño del fin de barra de error
    plt.bar(x=bins[:-1], height=hist, width=np.diff(bins), yerr=error, 
            edgecolor='black', ecolor=ecolor, capsize=2, alpha=alpha, align= 'edge',label = label)    
    return ax


def funciones_montecarlo(a, b, f, n):
    
    '''Dada una densidad de probabilidad, esta función devuelve un array de
    numpy con os datos necesarios para realizar el histograma de dicha 
    distribución mediante el método de montecarlo. Es necesario pasar como
    argumento el intervalo que queremos muestrar (a y b) y el número de puntos
    que queremos utilizar para el montercarlo, denotado como n.'''
    
    x_m= scipy.optimize.fmin(lambda x: -f(x), 0)[0]
    f_m = f(x_m)

    y = np.random.uniform(0, 1, n) 
    z = np.random.uniform(0, 1, n)
    
    u  = a + (b - a) * y
    v = f_m * z
    x = u[v<=h(u)]
    
    return x, v
```

#### Ejercicio a

En este ejercicio, se pide obtener la distribución de Cacuhy a partir de una distribución uniforme [0,1]. En el ejercicio 5 de la guía 3 vimos que basta con aplicarle la transformación $tan(\theta)$ a la mustra de una distribución uniforme en el intervalo $[-\frac{\pi}{2}, \frac{\pi}{2}]$. Para convertir el intervalo [0,1] utilizamos la transformación $a + (b − a)$ con $a = -\frac{\pi}{2}$ y $b = \frac{\pi}{2}$. A continuación se presenta el histograma de la muestra transformada en comparación con la distribución teórica de Cauchy.

```python
n = int(1e4)

a = np.random.uniform(0,1,n)
u  = -np.pi/2 + np.pi * a
b = np.tan(u)

plt.figure(figsize=(7 * 1.78, 7))
plt.grid(True, axis='y')
plt.xlim(-25,25)
plt.title('Histograma construído a partir del método de cambio de variables', size = 20)
plt.xlabel('Valores generados',size = 15)

t = np.linspace(-25,25,int(1e4))
plt.plot(t, cauchy.pdf(t), 'g-', linewidth = 4)

long = int(np.sqrt(len(b)))
ax = plot_histogram(b, np.linspace(-25, 25, long), alpha = 0.7)

plt.legend(loc='best')
plt.show()
```

#### Ejercicio b
Ahora utilizaremos el método Montecarlo para obtener el histograma de una distribución de Cauchy, asumiendo que conocemos la formula analítica de la distribución. Para ello generaremos varios pares de números aleatorios que corresponden a coordenadas en una región delimitada del plano. Los puntos $(u,v)$ se distribuyen de manera uniforme en los intervalos $u \in [-25,25]$ y $v \in [0, f_{max}]$ donde $f_{max}$ es el máximo de la función de distribución que queremos aproximar. Para cada par $(u*, v*)$, solo nos quedamos con los pares que cumplan con la condición $v* \leq f_{max}$. De los pares que cumplan dicha condición, realizamos un histograma de los valores de $u$. Se muestra el histograma comparándolo con la distribución de Cauchy teórica. 

Los errores en este histograma como en el anterior se tomaron poissoneanos, ya que la cantidad de datos y de bines se consideró lo suficientemente alta. Se estimaron los errores en las alturas en el histograma como la raiz cuadrada de la cantidad de datos en cada bin (estimación de Pearson).

```python
n = int(1e7)

#Cauchy
h = lambda x: 1/(np.pi*(1+x**2))

a,_ = funciones_montecarlo(-100,100, h, n)


plt.figure(figsize=(7 * 1.78, 7))
plt.grid(True, axis='y')
plt.xlim(-25,25)
plt.title('Histograma construído a partir del método de Montecarlo', size = 20)
plt.xlabel('Valores generados',size = 15)


t = np.linspace(-100,100,int(1e4))


plt.plot(t, cauchy.pdf(t) , 'g-',  linewidth = 4, label='Curva teórica')
long = np.sqrt(len(a))
ax = plot_histogram(a, np.linspace(-100,100, int(long)), alpha = 0.7)




plt.grid(True, axis='y')
plt.legend(loc='best')

plt.show()


print('Porcentaje de números generados que sobreviven: {}'.format(len(a)*100/n))
```

#### Ejercicio b
Se pide graficar la banda de error para cada $y_a$. Para calcularla utilizamos la fórmula de propagación de errores que vimos en la cursada, propagando los parámetros $a_1$ y $a_2$ como variables aleatorias.

$$Var(y_{a}) = \sum_{i} (\frac{\partial y_{a}}{\partial x_{i}})^2 \cdot \sigma_{i}^2 = (\frac{\partial y_{a}}{\partial a_{1}})^2 \cdot \sigma_{a_{1}}^2+ (\frac{\partial y_{a}}{\partial a_{2}})^2 \cdot \sigma_{a_{2}}^2+\frac{\partial y_{a}}{\partial a_{1}}\frac{\partial y_{a}}{\partial a_{2}}\cdot cov(a_1,a_2)$$

Realizando esta cuenta para cada valor de $y_a$ correspondiente a un dado $x_a$ (se utilizó una función definida especialmente llamada ¨propagar_f¨), se construyó una franja de errores para $y_a$ como se muestra en el siguiente gráfico. Se observa que la franja posee su mínimo aproximadamente en el valor de x que coincide con el promedio de los datos $\bar{x}$. La banda de error ¨pivotea¨ alrededor de ese valor de forma tal que una pendiente mas grande implicaría una ordenada al origen mas pequeña (errores correlacionados negativamente).

```python
t = np.linspace(0,5,1000) #Definimos un intervalo cuasi continuo

t_y = np.zeros(len(t))
var_t_y = np.zeros(len(t))
for i in range (len(t)):
    t_y[i], var_t_y[i] = propagar_f(t[i], covarianza=True)

t_ey = np.sqrt(var_t_y)

valor_buscado = t[t_ey==min(t_ey)]

print('El valor de x_a donde el error en y es mínimo es {}' .format(*valor_buscado))
print('El valor medio de x es {}' .format(np.mean(x)))
    
plt.figure(figsize=(7 * 1.78, 7))    
plt.title('Ajuste por cuadrados mínimos. Franja de errores en $y$ suponiendo correlación entre parámetros'
          , size = 20)
plt.xlim(0,5)
plt.ylim(0,8)
plt.xlabel('$x$ data', fontsize=15)
plt.ylabel('$y$ data', fontsize=15)
plt.grid(True)


plt.errorbar(x, y, yerr = y_error, fmt = 'r.', label = 'datos')
plt.plot(t,f(t, *popt), label = 'Ajuste', color = 'orange', linewidth = 2)
plt.fill_between(t, t_y-t_ey, t_y+t_ey,
    alpha=0.2, edgecolor='#1B2ACC', facecolor='#089FFF',
    linewidth=4, linestyle='dashdot', antialiased=True,  label= 'Banda de errores')

plt.legend(loc = 'best')
plt.show()
```
