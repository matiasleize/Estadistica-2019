Primero importamos las librerias y definimos las funciones que vamos a utilizar

```python
import numpy as np
import scipy as sp
from scipy.stats import binom, poisson
from matplotlib import pyplot as plt
import pandas as pd
```

```python
def plot_histogram(data, bins, ticks=5, xlabel='X', ylabel='Histograma', 
                   normed=True, color='', ecolor='' ,ax = None, alpha = None,
                   density = True):
    N = len(data)
    num_bins = len(bins)
    xmin = min(bins)
    xmax = max(bins)

    if ax is None:
        ax = plt.gca()
        
    if alpha is None:
        alpha = 1

    hist, _ = np.histogram(data, bins=bins, density=density)

    df = pd.DataFrame({'histograma': hist, 'error': np.sqrt(hist/N)})

    ax = df.plot.bar(y='histograma', yerr='error', width=1, edgecolor='k', ecolor='red', capsize=3, label=ylabel, 
                     ax=ax, color ='blue', alpha = alpha)
    
    plt.grid(True, axis = 'y')
    plt.xticks(np.linspace(0, num_bins - 2, ticks), np.linspace(xmin, xmax, ticks), rotation='horizontal');
    plt.xlabel(xlabel);
    return ax
```

Defino una función que toma como imput un número entre cero y uno (p) y el número natural N. Se generan N números aleatorios con distribución uniforme, listados en la lista n. De los N numeros, me quedo por un lado con los que sean menor al valor de p, listados en la lista n_p, y por otro con los números menores a su complemento definido como q = (1-p), que son listados en la lista n_q. Definimos la probabilidad de p y de q como el cociente entre la longitud de los vectories filtrados n_p y n_q (casos favorables), dividido por la cantidad de números totales generados N.

Para mostrar que el cociente obtenido coincide con la probabilidad p (es análogo para q), vamos a barrer distintos valores de p y graficaremos la respuesta de la funcion prob_uniforme en función de los valores de entrada. Se observa que a medida que se realizan mayor cantidad de simulaciones, las fluctuaciones alrededor de la recta con pendiente 1 se vuelven despreciables.

```python
def prob_uniforme(p, N=10e6):
    q = (1-p)
    n = np.random.uniform(0,1,int(N))
    n_p = n[n<=p]
    n_q = n[n<=q]
    prob_p = len(n_p)/len(n)
    prob_q = len(n_q)/len(n)
    return prob_p, prob_q


t = np.linspace(0,1,1000) #este vector auxiliar barre mil valores entre cero y uno que se utilizarán como imput
                            #de la función

#Falta hacer varios subplots
for N in [1000,10000,100000]:
    ps = np.zeros(len(t))
    qs = np.zeros(len(t))
    for i in range(len(t)):
        ps[i], _ = prob_uniforme(t[i],N)
    plt.figure(figsize=(9.7,6))
    plt.title('Prob_p vs Número p')
    plt.xlabel('Imput')
    plt.ylabel('Output')
    plt.grid(True)
    plt.plot(t,ps, label = 'Números simulados = {}'.format(N))
    plt.legend(loc = 0)
    plt.show()    
```

A partir de lo anterior, defino una función para generar n numeros aleatorios con probabilidad p de la siguiente forma

```python
def exp_bernoulli(N,p):
    '''Esta función toma como parametros el número de experimentos
    realizados y la probabilidad de tener éxito en cada uno de ellos.
    Devuelve el numero de exitos que hubo en el total de experimentos.'''
    n = np.random.uniform(0,1,N)
    n_p = n[n<=p] 
    cant_exitos = len(n_p)
    return cant_exitos

#A modo de ejemplo:
e_1= exp_bernoulli(N = int(10e6),p = 0.5)
print(e_1)
```

## Ejercicio 3

Ahora la fuente posee una intensidad media $I = 15 \frac{fotones}{seg}$. EL objetivo es simular los fotones emitidos solo en un $\Delta t = 1seg$. Vamos a subdividir este intervalo en $m$ intevalos iguales de longitud $dt$, de la forma $\Delta t = m \cdot dt$. Si dt es suficientemente pequeño, la probabilidad de que emita más de un fotón en un $dt$ es despreciable. Vamos a utilizar la función definida en el ejercicio 1 ya que en cada $dt$ consideramos un experimento de Bernoulli, esto es, la probabilidad de emitir un foton es fija ($p = I \cdot dt$) y solo hay probabilidad de éxito o fracaso (se emite o no se emite un fotón). La suma de todos los fotones emitidos en $\Delta t$ equivalen al numero de éxitos al tirar una moneda $m$ veces. Si realizamos este experimento un número suficientemente grande de veces (n_exp = 1000), condierando que la probabilidad de tener éxito es baja, la distribución binomial que obtendremos tenderá a parecerse a una distribución Poissoneana. $$\mu = I \cdot \Delta t = 15 \frac{fotones}{seg} \cdot 1 seg = 15 fotones$$

Comparamos los resultados con una distribución de Poisson teórica con ($\mu = 15$). Como en el punto anterior, por una cuestión de visualización se unieron los puntos con líneas, sin olvidar que tratamos con distribuciones discretas.  

```python
#Datos del problema
I = 15
m = 1000 #Lo que divido mi delta t
Dt = 1
dt = Dt/m

n_exp = 1000 #Cantidad de veces que se realiza el experimento

# Para un dt
p=I*dt
n_dt = m #n de la binomial

data = np.zeros(n_exp) #Hago mil experimentos
for j in range(n_exp):  
    data[j] = exp_bernoulli(m,p) #Hago m tiradas de moneda donde puedo tener un exito o un fracaso.

plt.figure(figsize=(9.7,6))
plt.title('Ejercicio 3')
plot_histogram(data,bins=np.arange(np.sqrt(len(data))))

Y = poisson(15) #Distribución de Poisson con mu = 15
t = np.arange(0,200)
plt.plot(t, Y.pmf(t), 'm.-', linewidth=4, ms=15, label ='Distribución teórica')
plt.xlabel('Número de fotones emitidos')
plt.legend(loc = 0)
plt.show()
```

## Ejercicio 5

En este ejercicio se pide que en vez de componer la distribución binomial del detector con la distribución de Poisson de la fuente, se encuentre un $\mu$ efectivo llamado $\tilde{\mu}$ para representar la distribución de todo el sistema.
Para ello podemos pensar que la probabilidad de detección de la fuente será la probabilidad de detección si la fuente ha emitido (que es la eficiencia $\epsilon = 0.75$) multiplicada por la probabilidad de que la fuente emita (que como observamos antes es Poissoneana), por lo que obtenemos que $$\tilde{\mu} = \epsilon \cdot \mu = 0.75 \cdot 15 = 11.25$$. Si realizamos un histograma de la nueva distribución comparándola con la curva teórica, esta se solapa con los errrores del histograma como ocurrió en el ejercicio anterior.

```python
#Datos del problema
I = 15
m = 1000 #Lo que divido mi delta t
Dt = 1
dt = Dt/m

n_exp = 1000 #Hago mil experimentos

# Para un dt
p=I*dt
n_dt = m #n de la binomial

eficiencia_detector = 0.75
prob_total = p*eficiencia_detector 

ev_detectados = np.zeros(n_exp) #Hago mil experimentos
for j in range(n_exp):  
    ev_detectados[j] = exp_bernoulli(m,prob_total)
    
plt.figure(figsize=(9.7,6))
plt.title('Ejercicio 5')
ax = plot_histogram(ev_detectados, np.arange(np.sqrt(len(ev_detectados))))

Y = poisson(eficiencia_detector*15) #Distribución de Poisson con mu = 15 por la eficiencia del detector
plt.plot(t, Y.pmf(t), 'k.-',ms=15, linewidth=4, label ='Distribución teórica')
plt.xlabel('Número de fotones detectados')
plt.legend(loc = 0)
plt.show()
```

## Ejercicio 6

Supongamos que tenemos N datos que queremos graficar en un histograma con $n$ bines. Para un bin, el evento 'dato cae dentro de un bin' es un experimento de Bernoulli. Ahora bien, la probabilidad de tener K éxitos (esto es, que en un bin caigan K datos) tiene una distribución binomial. En un histograma, en general la cantidad de datos es grande comparado con la cantidad de bines que posee el histograma (la probabilidad de que un dato caiga en un determinado bin es muy baja). Por esta razón, la distribución de probabilidad para la variable aleatoria 'Número de datos que caen dentro de un mismo bin' tiene distribución Poissoneana. Bajo este razonamiento, el error que se le debe asignar a dicho evento estará dado por el error de la distribución de Poisson $$\sigma_{poisson} = \sqrt{\mu}$$, donde $\mu$ será la altura de cada bin con su correspondiente normalización. Este es el error que se utilizó para realizar los histogramas de los puntos anteriores.

Pueden existir discrepancias entre los datos en el histograma y la distribución teórica. Esto se debe a que la cantidad de datos simulados no es suficientemente alta para que esto no ocurra. Para mejorar este aspecto una solución posible es aumentar la cantidad de experimentos realizados. De esta manera, el histograma tenderá a parecerse a la distribución teórica.
